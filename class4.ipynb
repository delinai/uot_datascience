#### Column Operations

```python
# when we filter a column we need to check for a condition and put that condition inside a column slice
# first, how do we isolate a column?
# this process works similar to dictionaries; select the column name, put inside square brackets, and the corresponding values are returned
df['province']
```

```python
# to check the object type
type(df['province'])
```

```python
# to isolate multiple columns, we need to pass a list of columns to isolate
df[['cust_name','province']]
```

```python
# to filter, first we need to create our filter. For example, let's find customers who live in ON. First, we need to create the ON filter
df['province'] == 'ON'
```

```python
# in order to return the full data frame with this condition, we place the condition inside the square brackets AS IF we are isolating a column
df[df['province'] == 'ON']
```

```python
# create a filter to find customers with more than 2 products
df[df['num_products'] > 2]
```

```python
# how to filter using multiple conditions - you need to isolate each condition inside a set of ROUND brackets (). we will use "&" to specify that both conditions
# should be met
df[(df['province']=='ON') & (df['num_products'] > 2)]
```

```python
# try it yourself! can we add a filter to specify ONLY customers in ON with more than 2 products who are WEATH customer category
df[(df['province']=='ON') & (df['num_products'] > 2) & (df['cust_category']=='wealth')]
```

```python
# we can do OR conditions using the pipe symbol "|"df
df[(df['province']=='ON') | (df['province']=='BC')]
```

```python
#Let's say we wanted to find customers in ON or BC, with more than 2 products - without proper brackets the function will combine filters with "&" and separate filters with "|"
df[(df['province']=='ON')|(df['province']=='BC')&(df['num_products']>2)]
```

```python
# we need additional brackets to bucket filters together
df[((df['province']=='ON')|(df['province']=='BC')) & (df['num_products']>2)]
```

```python
# column operations - sorting columns
df.sort_values('num_products') #using the sort_values function, specify the column to sort
```

```python
# to sort in descending order
df.sort_values('num_products', ascending=False)
```

```python
# to make the change permanent, we can add inplace=True
df.sort_values('num_products', ascending=False, inplace=True)
```

```python
# find the top 5 customers with the most products
top_5 = df.sort_values('num_products', ascending=False).head(5)
```

#### Aggregation Operations

```python
# aggregate a particular column
df['num_products'].mean() # apply summary statistics right at the end of a filter
```

```python
# sum total number of products across all customers
df['num_products'].sum()
```

```python
# we can layer aggregation functions with filters >> can we find the average number of products for people in ON?
df[df['province']=='ON']['num_products'].mean()
```

```python
# you can also save a separate object with the ON customers
cust_on = df[df['province']=='ON']
```

```python
# calculate the average as usual
cust_on['num_products'].mean()
```

```python
# let's calculate the average number of products for people in BC
df[df['province']=='BC']['num_products'].mean()
```

```python
# isolate both the number of products AND the len_relationship for customers in BC; then calculate the average of both
df[df['province']=='BC'][['num_products','len_relationship']].mean()
```

```python
# using groupby, pivot_table, crosstab we can aggregate the entire dataframe at once
# the logic of groupby, is that we start by specifying the column we want to aggregate data by - let's try with province
df.groupby('province') # this statement simply creates a grouped object
```

```python
# we need to evolve the function to tell Python, WHICH columns to aggregate based on Province, and HOW
df.groupby('province')[['num_products','len_relationship']].mean() # instead of filtering separately for each province, we can group by province to get the results for all
```

```python
# we can use a variety of aggregation functions, including sum(), min(), max(), std(), mean(), median()
df.groupby('province')[['num_products','len_relationship']].sum()
```

```python
# here is an example with standard deviation
df.groupby('province')[['num_products','len_relationship']].std()
```

```python
# maybe i want the average number of products and the minimum length of relationship - we can use aggregate()
df.groupby('province')[['num_products','len_relationship']].aggregate(['min','mean'])
```

```python
# i can also be more specific using a dictionary, to speficy how EACH column should be aggregated
df.groupby('province')[['num_products','len_relationship']].aggregate({'num_products':'mean','len_relationship':'min'})
```

```python
# how would we aggregate by customer category in addition to province? simply pass a LIST of columns in the groupby() function

df.groupby(['province','cust_category'])[['num_products','len_relationship']].aggregate({'num_products':'mean','len_relationship':'min'})
```

```python
# we can use pivot_table() to organize data with 3 parameters > index, values, columns

df.pivot_table(index=['province'],values=['num_products','len_relationship'],columns='cust_category',aggfunc={'len_relationship':'mean','num_products':'sum'})
```

```python
# to get the same view as groupby(), use the index and values parameters only
df.pivot_table(index=['province','cust_category'],values=['num_products','len_relationship'],aggfunc={'len_relationship':'mean','num_products':'sum'})
```

```python
# crosstab() -- can support aggregation across columns, using margins=True
pd.crosstab(df['province'],df['cust_category'],df['len_relationship'], aggfunc='sum', margins=True)
```
