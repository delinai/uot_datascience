### Class 5 - working with the Titanic Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
```

```python
# check structural information about the data

titanic.info()
```

```python
# isolate only the columns that we need
titanic_clean = titanic[['survived','pclass','sex','age','sibsp','parch','fare','embarked','deck']]
```

```python
titanic_clean.head()
```

```python
titanic_clean.info()
```

```python
# types of missing data
# MCAR - data missing completely at random - no reason why, every data point has equal likelihood of being missing
# MAR - data missing at random - there is a reason why the data point is missing
# MNAR - data missing not at random - data is missing for a particular reason

# if less than 10% of data is missing, drop it
# if between 10-40% is missing, fill it in
# if more than 40% is missing, drop the column

# checking the age column - about 80% of the data is there, and we have 20% of data missing
714/891
```

```python
titanic_clean['age'].mean()
```

```python
# fillna() function from pandas helps us fill in missing data
# example of filling in with the average of the whole column

titanic_clean['age'].fillna(titanic_clean['age'].mean())
```

```python
# check if the average by sex is different

titanic_clean.groupby('sex')['age'].mean()
```

```python
titanic_clean.groupby('pclass')['age'].mean()
```

```python
# the first part of the function isolates the column we are filling in the missing data for - in our case: titanic_clean['age']
# the second part of the function, fillna() isolates just the missing values
# the first part of the function inside fillna() groups the data by pclass and isolates the ages
# the transform function replaces missing values NaN with the 'mean' of the grouped ages

titanic_clean['age'].fillna(titanic_clean.groupby('pclass')['age'].transform('mean'))
```

```python
# to make the change permanent, we need to set the age column equal to this transformation

titanic_clean['age'] = titanic_clean['age'].fillna(titanic_clean.groupby('pclass')['age'].transform('mean'))
```

```python
# the regular groupby().mean() scenario calculates the mean value by group

titanic_clean.groupby('pclass')['age'].mean()
```

```python
# the transform function replaces every value with the mean, based on the pclass

titanic_clean.groupby('pclass')['age'].transform('mean')
```

```python
# since we don't want to replace every value, but only the missing values, we wrap the above function inside fillna()
titanic_clean.head()
```

```python
titanic_clean.info()
```

```python
# the deck column has too much missing data so we will drop it
# the drop function can drop columns or rows
# for columns, we specify the column name, axis=1, and inplace=True to make the change permanent

titanic_clean.drop('deck', axis=1, inplace=True)
```

```python
titanic_clean.info()
```

### Initial hypotheses about the drivers of survival

1. We think that women in first class had a higher likelihood of survival
2. We think that women with children had a higher likelihood of survival

We will start with univariate analysis (each variable by itself), followed by bi-variate and multi-variate.

```python
# check the survived column
titanic_clean.groupby('survived').count()
```

```python
# create a plot with pandas plot
fig, (ax) = plt.subplots(1,1, figsize=(4,4))

# create a plot with pandas plot
titanic_clean.groupby('survived')['pclass'].count()\
.plot(ax=ax, kind='bar', title='Survival Counts', ylabel='Count of Passengers')

ax.annotate(titanic_clean.groupby('survived')['pclass'].count().loc[0],(0,300), ha='center', color='white', fontsize=15)
ax.annotate(titanic_clean.groupby('survived')['pclass'].count().loc[1],(1,200), ha='center', color='white', fontsize=15)
```

```python
titanic_clean.groupby('survived')['pclass'].count().loc[0]
```

```python
# we can create a loop for the annotation function

# create a plot with pandas plot
fig, (ax) = plt.subplots(1,1, figsize=(4,4))

# create a plot with pandas plot
titanic_clean.groupby('survived')['pclass'].count()\
.plot(ax=ax, kind='bar', title='Survival Counts', ylabel='Count of Passengers')

for i in range(2):
    ax.annotate(titanic_clean.groupby('survived')['pclass'].count().loc[i],(i,(titanic_clean.groupby('survived')['pclass'].count().loc[i]/2)), ha='center', color='white', fontsize=15)

# add the word "survivors" above the survivors bar
ax.annotate('survivors',(1,400), ha='center')

ax.set_xticks(labels=['no','yes'], ticks=[0,1], rotation=360)
```

```python
# let's plot a stacked bar chart that shows survivors by sex

titanic_clean.groupby(['survived','sex'])['pclass'].count()#.plot(kind='bar', stacked=True)
```

```python
# we need to convert a multi-index dataframe into a matrix to create a stacked bar chart
titanic_clean.groupby(['survived','sex'])['pclass'].count().unstack()
```

```python
# create a plot with pandas plot
fig, (ax) = plt.subplots(1,1, figsize=(4,4))

# create a plot with pandas plot
titanic_clean.groupby(['survived','sex'])['pclass'].count().unstack().plot(kind='bar',
                                                                            stacked=True, ax=ax, title='Survivors by Sex',
                                                                            ylabel='Count of survivors')

for i in range(2):
    ax.annotate(titanic_clean.groupby(['survived','sex'])['pclass'].count().unstack().iloc[i,0],(i,(titanic_clean.groupby(['survived','sex'])['pclass'].count().unstack().iloc[i,0]/2)), ha='center', color='white', fontsize=15)
    ax.annotate(titanic_clean.groupby(['survived','sex'])['pclass'].count().unstack().iloc[i,1],(i,(titanic_clean.groupby(['survived','sex'])['pclass'].count().unstack().iloc[i,1]/2)), ha='center', color='white', fontsize=15)

ax.set_xticks(labels=['no','yes'], ticks=[0,1], rotation=360)

```
