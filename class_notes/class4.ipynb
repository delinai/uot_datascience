```python
# how can we find customers who are in the personal_banking category?

customers[customers['cust_category'] == 'personal_banking']
```

```python
# we can filter by two conditions by separating the conditions with an operator - & = and, | = or
# let's find customers in Ontario who are personal_banking customers
# we need a set of round brackets around each condition

customers[(customers['cust_category']=='personal_banking') & (customers['province']=='ON')]
```

```python
# let's try an OR statement using |
# let's find customers who are personal_banking customers OR live in BC

customers[(customers['cust_category'] == 'personal_banking') | (customers['province'] == 'BC')]
```

```python
# let's find customers who are high value > this is defined as either customers who have been with the bank for longer than 3 years
# OR customers who have at least 3 accounts

hvl_cust = customers[(customers['len_relationship'] > 3)|(customers['num_products'] >= 3)]
hvl_cust
```

```python
# if we have multiple conditions for the same column, we can use isin()
# let's find all customers in ON and BC
# every time we have multiple conditions, we must pass a list or tuple

customers[customers['province'].isin(['BC','ON'])]
```

```python
# how to aggregate data
# let's say we want to see the average number of banking products customers in Ontario have
# one way to do this is to filter

customers[customers['province']=='ON']['num_products'].mean()
```

```python
# what if our question was: what is the avg number of products in each province?
# a traditional way would be to use a loop

# step 1: find unique values for province
provinces = customers['province'].unique()

# step 2: iterate through the dataframe using a loop

for i in provinces:
    print(i, customers[customers['province']==i]['num_products'].mean())
    
```

```python
# the unique function will return the unique values within a column
customers['province'].unique()
```

```python
# the last (and best) option is to use groupby()

customers.groupby('province')['num_products'].mean()
```

```python
# let's find the average length of relationship, AND the average number of products

customers.groupby('province')[['num_products','len_relationship']].mean() # use double square brackets to pass a list of columns to aggregate
```

```python
# we can pass a list of columns to groupby() for grouping by multiple conditions
# this is called a multi-index dataframe

customers.groupby(['province','cust_category'])[['num_products','len_relationship']].mean()
```

```python
# to view multiple metrics, we can use the aggregate function
# we can pass a LIST of aggregation metrics in the aggregate function to return for each column

customers.groupby(['province','cust_category'])[['num_products','len_relationship']].aggregate(['mean','sum'])
```

```python
# we can pass a dictionary with specific instructions for each column
# add the rename() function to rename the columns to reflect the aggregation
# add the backslash in order to move part of the function to a new line for better readibility

customers.groupby(['province','cust_category'])[['num_products','len_relationship']]\
    .aggregate({'num_products':'mean','len_relationship':'max'})\
        .rename(columns={'num_products':'mean_num_products','len_relationship':'max_rel'})
```

```python
# let's say we want to see the average num_products by province and category in a matrix; we can use unstack()
# to unstack the multi-index

customers.groupby(['province','cust_category'])['num_products'].mean().unstack()
```

```python
# merging and joining dataframes
# import the transactions file

transactions = pd.read_csv('https://raw.githubusercontent.com/delinai/uot_datascience/main/transactions.csv')
```

```python
transactions.head()
```

```python
customers.head()
```

```python
# we can use the join() function or the merge() function
# join() requires both dataframes to have the SAME index; it joins on the index directly

# Step 1: set the index of the transactions table as customer_id

transactions.set_index('customer_id',inplace=True)
```

```python
# step 2: create a new DF and join the two together
joined_data = customers.join(transactions)
```

```python
joined_data.head(2)
```

```python
# let's clean up the data - first, rename the transaction id column from "id" to "txn_id", then convert to integer from float

# Step 1: rename column
joined_data.rename(columns={'id':'txn_id'}, inplace=True)

```

```python
# check for missing data - next week we will learn how to fill in missing data, and address the data type
joined_data.info()
```

```python
# we can also use merge() to connect two dataframes; this function does not require both frames to have the same index
# left parameter determines the table on the left
# right parameter determines the table on the right
# right_on identifies the common column which we should merge / lookup on
# left_index tells Python to merge on that index with the right_on parameter, and it keeps the index inplace

merged_data = pd.merge(left=customers, right=transactions, right_on='customer_id', left_index=True)
```
