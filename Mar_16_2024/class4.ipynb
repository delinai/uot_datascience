```python
# groupby - aggregating data / statistics based on group
customers_df
```

```python
# how many customers fall into a category (e.g., ON, hnw)? >> agg function = COUNT
# what is the average length of relationship by cust category? >> agg function = MEAN
# how many customers fall into a hnw category by province? >> agg function = COUNT
# % of customers by province? >> agg function = COUNT / Total
```

```python
# how many customers in each province? agg function = COUNT
customers_df.groupby('province')['cust_name'].count()
```

```python
# what is the average length of relationship by province? agg function = mean()
customers_df.groupby('province')['len_relationship'].mean()
```

```python
# we may have a more specific question - like # of customers by province AND category
# pass a list of categories to group by, in the groupby() statement
customers_df.groupby(['province','cust_category'])['cust_name'].count()
```

```python
list_object = [1,2]
```

```python
# we can also use aggregate() and pass a list of aggregations to apply to a single column
customers_df.groupby(['province','cust_category'])['len_relationship'].aggregate(['mean','count'])
```

```python
# we can also aggregate more than one column
# apply the same aggregation method (e.g., mean()) to calculate the avg num of products and avg length of relationship
customers_df.groupby(['province','cust_category'])[['num_products','len_relationship']].mean()
```

```python
# we can also apply custom aggregations to each column using the aggregate() function with a dictionary 
# what is the highest length of relationship by province and category, and the avg # of products

customers_df.groupby(['province','cust_category'])[['num_products','len_relationship']]\
    .aggregate({'num_products':'mean','len_relationship':'max'})
```

```python
# to change the names of the columns, store the analysis in a new object and rename

analysis = customers_df.groupby(['province','cust_category'])[['num_products','len_relationship']]\
    .aggregate({'num_products':'mean','len_relationship':'max'})
analysis.columns = ['avg num products','max len relationship']
```

```python
analysis
```

```python
# pandas will produce aggregation labels for each column, if you have multiple aggregations for each column
# pass a list of aggregation functions in the dictionary to do this

customers_df.groupby(['province','cust_category'])[['num_products','len_relationship']]\
    .aggregate({'num_products':['mean','max'],'len_relationship':['mean','max']})
```

```python
transactions = pd.read_csv(r'/Python For Data Science/transactions.csv')
```

```python
transactions.head()
```

```python
customers_df.reset_index(inplace=True)
```

```python
# the rename function can help us rename specific columns
customers_df.rename(columns={'id':'customer_id'}, inplace=True)
```

```python
# when you pass a list of column names to the df.columns attribute, you need to include every single column name
# otherwise, if you only want to change the name of one column, use df.rename()
# example: df.columns = [] vs. df.rename(columns={})
```

```python
# rename the id in the transactions table to txn_id so it's more clear
transactions.rename(columns={'id':'txn_id'}, inplace=True)
```

```python
# to join tables together, we can use merge() or join()
# the merge() function is a library-level function and is more flexible, doesn't care about column names, whether one is an index or not
# the join() function is a class method, and does care about names/index; less flexible, but more efficient

merge_df = pd.merge(customers_df, transactions, on = 'customer_id')
```

```python
merge_df.head()
```

```python
# the join function requires that both tables you are joining have the same index
joined_df = customers_df.set_index('customer_id').join(transactions.set_index('customer_id'))
```

```python
joined_df.head()
```

```python
# you can continue analyzing the joined tables using the groupby methods or other functions
joined_df.groupby('cust_category')['txn_total'].sum()
```

```python
# use describe() to see descriptive stats
joined_df.describe(include='all')
```

```python
# to save the transformed file, pick the df you want to save, use to_csv() and specify the file path + file name at the end
joined_df.to_csv('/Python For Data Science/customers_and_transactions.csv')
```
